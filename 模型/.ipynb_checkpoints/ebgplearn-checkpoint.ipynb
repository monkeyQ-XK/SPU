```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from gplearn.genetic import SymbolicRegressor
import sklearn
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sympy import *
from sklearn.utils.random import check_random_state
import graphviz
import time
from sklearn.model_selection import train_test_split  #分割数据集
from sklearn.metrics import mean_squared_error,r2_score,mean_absolute_error   #模型评价
import warnings
warnings.filterwarnings("ignore")
```

```python
converter = {
    'add': lambda x, y : x + y,
    'sub': lambda x, y : x - y,
    'mul': lambda x, y : x*y,
    'div': lambda x, y : x/y,
    'sqrt': lambda x : x**0.5,
    'log': lambda x : log(x),
    'abs': lambda x : abs(x),
    'neg': lambda x : -x,
    'inv': lambda x : 1/x,
    'max': lambda x, y : max(x, y),
    'min': lambda x, y : min(x, y),
    'sin': lambda x : sin(x),
    'cos': lambda x : cos(x),
    'pow': lambda x, y : x**y,
}
```

```python
df=pd.read_csv('Eb.csv')
df=df.drop(['Unnamed: 0'],axis=1)
X=df.iloc[:,0:52]
y=df.iloc[:,52]
y
X=X.values
y=y.values.ravel()
```

```python
p_crossover1=[0.7]# np.arange(1,101,10
p_subtree_mutation1=[0.1]
p_hoist_mutation1=[0.1]
#p_point_mutation=
random_state1=np.arange(0,5,1)
random_state2=np.arange(0,1,1)
function_set = ['add', 'sub', 'mul', 'div',
                'sqrt', 'log', 'abs', 'neg', 'inv',
                'sin', 'cos','tan']
```

```python
from sklearn.preprocessing  import StandardScaler
from gplearn.genetic import SymbolicTransformer,SymbolicRegressor
```

```python
#import xgboost as xgb
best_score =0
for p_crossover in p_crossover1:
    for p_subtree_mutation in p_subtree_mutation1:
        for p_hoist_mutation in p_hoist_mutation1:
            #for p_point_mutation in p_point_mutation:
            for random_state1 in random_state1:
                for random_state2 in random_state2:
                    X_train,X_test= train_test_split(X, test_size=0.2,random_state =random_state1)
                    y_train,y_test = train_test_split(y, test_size=0.2,random_state = random_state1)
                    sc = StandardScaler()
                    X_train = sc.fit_transform(X_train)
                    X_test= sc.transform(X_test)
                    est_gp = SymbolicRegressor(population_size=5000,
                                               generations=50, stopping_criteria=0.001,parsimony_coefficient=0.001,
                                              function_set=function_set,
                                               p_crossover=p_crossover, p_subtree_mutation=p_subtree_mutation,
                                               p_hoist_mutation=p_hoist_mutation, p_point_mutation=1-p_crossover-p_hoist_mutation-p_subtree_mutation,
                                               verbose=1,
                                    random_state=random_state2)
                    est_gp.fit(X_train,y_train ) 
                    score=est_gp.score(X_test,y_test) 
                    #print(est_gp.score(X_train,y_train))#0.5967867578111098
                    #print(est_gp._program)
                    if score > best_score:
                        best_score = score
                        best_parameters = {'n_estimators': p_crossover, 'max_depth': p_subtree_mutation,'c':p_hoist_mutation,"d":random_state1,'e':random_state1
                                          ,"f":random_state2}
print("Best score: {:.5f}".format(best_score))
print("Best parameters: {}".format(best_parameters)) 
```

```python

X_train,X_test= train_test_split(X, test_size=0.2,random_state =5)
y_train,y_test = train_test_split(y, test_size=0.2,random_state = 5)
#from sklearn.preprocessing  import StandardScaler
#sc = StandardScaler()
#X_train = sc.fit_transform(X_train)
#X_test= sc.transform(X_test)
from gplearn.genetic import SymbolicTransformer,SymbolicRegressor
est_gp = SymbolicRegressor(population_size=5000,generations=112,stopping_criteria=0.0090,p_crossover=0.7,
                           p_subtree_mutation=0.1,p_hoist_mutation=0.05,p_point_mutation=0.1,function_set=function_set,init_depth=(4,8),
                           max_samples=0.5,verbose=2,parsimony_coefficient=0.0001,random_state=0)
est_gp.fit(X_train,y_train )
 
print(est_gp.score(X_test,y_test)) 
print(est_gp.score(X_train,y_train))#0.5967867578111098
print(est_gp._program)
```

```python
y_train_pred=est_gp.predict(X_train)
y_test_pred=est_gp.predict(X_test)
```

```python
plt.figure(figsize=(6,4)) # 指定图片的尺寸
plt.scatter(y_train,y_train_pred,color='red',marker='o',edgecolor='white',label ='Training data') #残差散点图，指定散点的颜色，标记(o为圆圈s为方块)，标签等设定
plt.scatter(y_test,y_test_pred,color='blue',marker='s',edgecolor='white',label ='Test data')
plt.ylabel('predicted ')
plt.xlabel('Experimental ')
plt.title('Comparison between the experimental and predicted ')
plt.legend(loc = 'upper left') #图例位置
plt.plot([1, 4], [1, 4],c='black')
plt.xlim([1,4]) # 设定x轴范围
plt.ylim(1,4)
plt.tight_layout() #自动调整子图参数，使之填充整个图像区域，可用于检查坐标轴标签、刻度标签以及标题的部分
#plt.savefig("C:\Users\Administrator\Desktop\test\ILCO2_SVM_pred.png")     #图片保存
plt.show()
```
